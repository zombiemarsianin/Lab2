## Loading data frame

```python
import pandas as pd
import os

os.chdir("/home/andriid/Documents/Labs/SPF/1")
print("enter folder")
path = input() + "/"

try:
    i = 0
    frames = {}
    for filename in os.listdir(path):
        if filename.endswith(".csv"):
            frames[filename] = pd.read_csv(path+"/"+filename, sep=",{1} *| {1,4}", index_col=False, engine='python', header=1)
except FileNotFoundError:
    print("error: no path with data provided!")
    
for filename in os.listdir(path):
    if filename.endswith(".csv"):
        print("\nframe: " + filename)
        print(frames[filename][:-1])
```
